```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import random
import math

plt.rcParams["figure.figsize"] = (20,10)
plt.style.use("ggplot")
```

```python
data = pd.read_csv("data/iris.csv", names=["sepal_length", "sepal_width", "petal_length", "petal_width", "label"])
data.head(2)
```

```python
plt.subplot(2,1,1)
sns.scatterplot(data=data, x="petal_length", y="petal_width", hue="label", cmap="tab10")
plt.subplot(2,1,2)
sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="label", cmap="tab10")
```

```python
# From the above graph we can conclude that information about the petals can help us classify iris-setosa from the order 2 classes
# We will hence create a logistic regression classifier that 
```

```python
data_subset = data[["petal_width","petal_length","label"]].copy()
data_subset = data_subset.replace("Iris-virginica","non_setosa").replace("Iris-versicolor","non_setosa")
```

```python

encoder = LabelEncoder()
new_labels = encoder.fit_transform(data_subset.label.values)
data_subset["encoded_labels"] = new_labels
```

```python
def bi_variable_custom_logistic_regression(X, y, w1=None, w2=None, b=None, alpha=0.001):
    if w1 is None:
        w1 = random.random()
    if w2 is None:
        w2 = random.random()
    if b is None:
        b = random.random()

    def _get_sigmoid_value(eq_value):
        return 1 / (1 + np.exp(-eq_value))

    def _get_eq_value(features_):
        return features_.dot([w1, w2]) + b

    def _get_performance():
        calculated_label = _get_sigmoid_value(_get_eq_value(X))
        cost = -(y * np.log(calculated_label) +
                (1 - y) * np.log(1 - calculated_label))
        average_cost = sum(cost) / len(cost)
        return calculated_label, average_cost
        
    iteration = 0
    while True:
        calculated_labels, average_cost = _get_performance()
        
        a_y = calculated_labels-y
        dw1 = X[:,0].dot(a_y)/len(X)
        dw2 = X[:,1].dot(a_y)/len(X)
        db = sum(a_y)/len(X)
        updated_w1 = w1-alpha*dw1
        updated_w2 = w2-alpha*dw2
        updated_b = b-alpha*db
        
        if math.isclose(w1, updated_w1, rel_tol=1e-05) and math.isclose(w2, updated_w2, rel_tol=1e-05) and math.isclose(b, updated_b, rel_tol=1e-05):
            break
        w1 = updated_w1
        w2 = updated_w2
        b = updated_b

    return w1, w2, b, _get_performance()[1]

def get_predictions(w1,w2,b,X):
    
    def _get_sigmoid_value(eq_value):
        return 1 / (1 + np.exp(-eq_value))

    def _get_eq_value(features_):
        return features_.dot([w1, w2]) + b
    
    return np.round(_get_sigmoid_value(_get_eq_value(X))).astype(int)
    
    
    
features = data_subset[["petal_width", "petal_length"]].values
actual_label = data_subset["encoded_labels"].values
x_train, x_test, y_train, y_test = train_test_split(features, actual_label)
w1, w2, b, performance = bi_variable_custom_logistic_regression(x_train, y_train)
predictions = get_predictions(w1,w2,b,x_test)
```

```python
print(classification_report(y_test, predictions))
```
